```python
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
TRAINING_DATA_DIR = '../../data/processed/training/'
TRAINING_DATA_DIR_32 = os.path.join(TRAINING_DATA_DIR, '32x32')
TRAINING_DATA_DIR_64 = os.path.join(TRAINING_DATA_DIR, '64x64')
```

```python
list_of_files = os.listdir(TRAINING_DATA_DIR_32)
```

```python
random.seed(521)
```

```python
index = random.randint(0, len(list_of_files))

img_32 = mpimg.imread(os.path.join(TRAINING_DATA_DIR_32, list_of_files[index]))
img_64 = mpimg.imread(os.path.join(TRAINING_DATA_DIR_64, list_of_files[index]))

fig = plt.figure()
a = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(img_32)
a.set_title('32x32')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
a = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(img_64)
imgplot.set_clim(0.0, 0.7)
a.set_title('64x64')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
```
